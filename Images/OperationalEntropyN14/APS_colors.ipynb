```python
import numpy as np
import matplotlib.pyplot as plt
import colors
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

## Define the Colors

```python
plot_colors = {}
plot_colors['red'] = ["#e85c47"]
plot_colors['blue'] = ["#4173b2"]
plot_colors['green'] = ["#7dcca4"]
plot_colors['orange'] = ["#fdbe6e"]

α = [0.8,0.5,0.1]

for col_name, color in plot_colors.items():
    for cα in α:
        color.append(colors.get_alpha_hex(color[0],cα))
```

## Print out the colors we are using

```python
import pprint
pprint.pprint(plot_colors)
```

## Plot them

```python
x = np.linspace(-1,1,10)

fig, axs = plt.subplots(2, 2, sharex=True, sharey=True,figsize=(12,12))

i = 0
for col_name, color in plot_colors.items():
    idx = np.unravel_index(i, (2,2))
    for j in range(3):
        axs[idx].plot(x,x**2+j,color=color[0], linewidth=1, markersize=8, markerfacecolor=color[j+1],marker='o', label=col_name)
    i += 1
    axs[idx].legend(loc='upper center', frameon=False, fontsize=18)

plt.savefig('color_test.pdf')
```
